```python
import cv2
import numpy as np

def detect_potholes(image_path):
    # Load the image
    image = cv2.imread(image_path)
    
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # cv2.imshow('GrayImage',gray)
    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    cv2.imshow('Blurred Image',blurred)
    # Apply adaptive thresholding to enhance the edges
    _, thresholded = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    cv2.imshow('Thresholded',thresholded)
    # Perform morphological operations to close gaps in between object edges
    kernel = np.ones((5, 5), np.uint8)
    closed = cv2.morphologyEx(thresholded, cv2.MORPH_CLOSE, kernel)
    cv2.imshow('closed',closed)
    # Find contours in the image
    contours, _ = cv2.findContours(closed.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Iterate over the contours and identify potential potholes
    potholes = []
    for contour in contours:
        perimeter = cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour, 0.02 * perimeter, True)
        
        # Check if the contour is roughly circular and has a minimum area
        if len(approx) >= 6 and cv2.isContourConvex(approx):
            area = cv2.contourArea(approx)
            circularity = (perimeter ** 2) / (4 * np.pi * area)
            if circularity > 0.8 and area > 100:
                potholes.append(approx)
    
    # Draw the detected potholes on the original image
    cv2.drawContours(image, potholes, -1, (0, 255, 0), 2)
    
    # Display the image
    cv2.imshow("Pothole Detection", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Path to the image you want to test
image_path = "pothole.jpg"

# Call the detect_potholes function
detect_potholes(image_path)
```
